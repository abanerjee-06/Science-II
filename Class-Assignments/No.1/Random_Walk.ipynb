```python
import random
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
stepnum = np.zeros(10000)
stepnum2 = np.zeros(10000)
#stepnum3 = np.zeros(100000)
#stepnum4 = np.zeros(100000)
calculated = np.zeros(100)
d1mean = 0
d2mean = 0
d1meansq = np.zeros(10000)
d2meansq = np.zeros(10000)
for i in range(10000):
    steps = 0
    d1 = 0
    d2 = 0
    for j in range(100):
        if d1 == d2:
            stepnum[steps] = stepnum[steps] + 1
        if (d1 == 0): 
            stepnum2[steps] = stepnum2[steps] + 1
        d1 = d1 + random.choice([-1,1])
        d2 = d2 + random.choice([-1,1])
        steps = steps + 1
#        if (d1 == 1) or (d2 == 1):
#            stepnum3[steps] = stepnum3[steps] + 1
#        if (d1 == -1) or (d2 == -1):
#            stepnum4[steps] = stepnum4[steps] + 1
        d1meansq[steps] = d1meansq[steps] + d1**2
        d2meansq[steps] = d2meansq[steps] + d2**2
    d1mean = d1mean + d1
    d2mean = d2mean + d2
for i in range(100):
    d1meansq[i] = d1meansq[i]/10000
    d2meansq[i] = d2meansq[i]/10000
    stepnum[i] = stepnum[i]/10000
    stepnum2[i] = stepnum2[i]/10000
d1mean = d1mean/10000
d2mean = d2mean/10000
for i in range(100):
    calculated[i] = math.factorial(2*i)/(((2**i)*math.factorial(i))**2)
```

```python
plt.plot(stepnum[:100])
plt.plot(calculated)
plt.xlabel('steps')
plt.ylabel('probability')
plt.title('Probability of meeting after n steps')
plt.legend(['Simulated','Analytically Calculated'])
```
